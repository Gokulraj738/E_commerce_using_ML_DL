```python
categorical_features_camera = ['s1_fact.order_payment_type', 'product_analytic_super_category', 
                               'product_analytic_category', 'product_analytic_sub_category', 
                               'product_analytic_vertical']

numeric_features_camera = ['units', 'sla', 'product_mrp']

# Create preprocessor for Camera Accessory
preprocessor_camera = ColumnTransformer(
    transformers=[
        ('num', StandardScaler(), numeric_features_camera),
        ('cat', OneHotEncoder(handle_unknown='ignore'), categorical_features_camera)
    ])

# Transform Camera Accessory data
X_train_camera_transformed = preprocessor_camera.fit_transform(X_train_camera)
X_test_camera_transformed = preprocessor_camera.transform(X_test_camera)

# Fit the Linear Regression model for Camera Accessory
model_camera = LinearRegression()
model_camera.fit(X_train_camera_transformed, y_train_camera)

# Evaluate the model
score_camera = model_camera.score(X_test_camera_transformed, y_test_camera)
print(f"R^2 Score for Camera Accessory: {score_camera}")
```

```python
from sklearn.metrics import mean_squared_error, r2_score
import numpy as np
```

```python
# Predictions on the test set for Camera Accessory

y_pred_camera = model_camera.predict(X_test_camera_transformed)

# Calculate R² Score
r2_camera = r2_score(y_test_camera, y_pred_camera)

# Calculate MSE
mse_camera = mean_squared_error(y_test_camera, y_pred_camera)

# Calculate RMSE
rmse_camera = np.sqrt(mse_camera)

# Calculate percentage of explained variance (same as R² score but in percentage)
explained_variance_camera = r2_camera * 100

# Print the results
print(f"Camera Accessory Model Performance:")
print(f"R² Score: {r2_camera}")
print(f"MSE: {mse_camera}")
print(f"RMSE: {rmse_camera}")
print(f"Explained Variance (%): {explained_variance_camera:.2f}%")
```

```python
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor
from sklearn.metrics import mean_squared_error, r2_score
import numpy as np
import xgboost as xgb

# Function to train and evaluate models
def train_and_evaluate_model(X_train, y_train, X_test, y_test, category_name):
    # Initialize models
    random_forest_model = RandomForestRegressor(n_estimators=100, random_state=42)
    xgboost_model = xgb.XGBRegressor(n_estimators=100, random_state=42)
    gbr_model = GradientBoostingRegressor(n_estimators=100, random_state=42)

    # Train the models
    random_forest_model.fit(X_train, y_train)
    xgboost_model.fit(X_train, y_train)
    gbr_model.fit(X_train, y_train)

    # Predictions
    y_pred_rf = random_forest_model.predict(X_test)
    y_pred_xgb = xgboost_model.predict(X_test)
    y_pred_gbr = gbr_model.predict(X_test)

    # Metrics for Random Forest
    r2_rf = r2_score(y_test, y_pred_rf)
    mse_rf = mean_squared_error(y_test, y_pred_rf)
    rmse_rf = np.sqrt(mse_rf)
    explained_variance_rf = r2_rf * 100

    # Metrics for XGBoost
    r2_xgb = r2_score(y_test, y_pred_xgb)
    mse_xgb = mean_squared_error(y_test, y_pred_xgb)
    rmse_xgb = np.sqrt(mse_xgb)
    explained_variance_xgb = r2_xgb * 100

    # Metrics for Gradient Boosting
    r2_gbr = r2_score(y_test, y_pred_gbr)
    mse_gbr = mean_squared_error(y_test, y_pred_gbr)
    rmse_gbr = np.sqrt(mse_gbr)
    explained_variance_gbr = r2_gbr * 100

    # Print the results
    print(f"\n\n========== {category_name} ==========")
    
    # Random Forest
    print(f"\nRandom Forest Performance for {category_name}:")
    print(f"R² Score: {r2_rf}")
    print(f"MSE: {mse_rf}")
    print(f"RMSE: {rmse_rf}")
    print(f"Explained Variance: {explained_variance_rf:.2f}%")
    
    # XGBoost
    print(f"\nXGBoost Performance for {category_name}:")
    print(f"R² Score: {r2_xgb}")
    print(f"MSE: {mse_xgb}")
    print(f"RMSE: {rmse_xgb}")
    print(f"Explained Variance: {explained_variance_xgb:.2f}%")
    
    # Gradient Boosting
    print(f"\nGradient Boosting Performance for {category_name}:")
    print(f"R² Score: {r2_gbr}")
    print(f"MSE: {mse_gbr}")
    print(f"RMSE: {rmse_gbr}")
    print(f"Explained Variance: {explained_variance_gbr:.2f}%")
    
# Camera Accessory Model
train_and_evaluate_model(X_train_camera_transformed, y_train_camera, X_test_camera_transformed, y_test_camera, "Camera Accessory")

# Home Audio Model
train_and_evaluate_model(X_train_home_audio_transformed, y_train_home, X_test_home_audio_transformed, y_test_home, "Home Audio")

# Gaming Accessory Model
train_and_evaluate_model(X_train_gaming_transformed, y_train_gaming, X_test_gaming_transformed, y_test_gaming, "Gaming Accessory")
```

```python
import numpy as np
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.optimizers import Adam
from sklearn.metrics import mean_squared_error, r2_score

# Function to build the model
def build_model(input_shape):
    model = Sequential()
    model.add(Dense(128, input_dim=input_shape, activation='relu'))  # First hidden layer
    model.add(Dense(64, activation='relu'))  # Second hidden layer
    model.add(Dense(32, activation='relu'))  # Third hidden layer
    model.add(Dense(1, activation='linear'))  # Output layer for regression
    model.compile(optimizer=Adam(learning_rate=0.001), loss='mean_squared_error')
    return model

# Function to train and evaluate the deep learning model
def train_and_evaluate_deep_model(X_train, y_train, X_test, y_test, category_name, epochs=50, batch_size=32):
    model = build_model(X_train.shape[1])
    
    # Train the model
    history = model.fit(X_train, y_train, validation_data=(X_test, y_test), epochs=epochs, batch_size=batch_size, verbose=1)
    
    # Make predictions
    y_pred = model.predict(X_test)
    
    # Evaluate model performance
    mse = mean_squared_error(y_test, y_pred)
    rmse = np.sqrt(mse)
    r2 = r2_score(y_test, y_pred)
    explained_variance = r2 * 100
    
    # Print the results
    print(f"\n\n========== {category_name} ==========")
    print(f"Deep Learning Model Performance for {category_name}:")
    print(f"R² Score: {r2}")
    print(f"MSE: {mse}")
    print(f"RMSE: {rmse}")
    print(f"Explained Variance: {explained_variance:.2f}%")

# Camera Accessory Model
train_and_evaluate_deep_model(X_train_camera_transformed, y_train_camera, X_test_camera_transformed, y_test_camera, "Camera Accessory", epochs=50, batch_size=32)

# Home Audio Model
train_and_evaluate_deep_model(X_train_home_audio_transformed, y_train_home, X_test_home_audio_transformed, y_test_home, "Home Audio", epochs=50, batch_size=32)

# Gaming Accessory Model
train_and_evaluate_deep_model(X_train_gaming_transformed, y_train_gaming, X_test_gaming_transformed, y_test_gaming, "Gaming Accessory", epochs=50, batch_size=32)
```
